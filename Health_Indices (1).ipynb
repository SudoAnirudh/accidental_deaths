```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dictionary with the provided data
data = {
    'Category': ['SC', 'ST', 'OD', 'Others', 'India'],
    'IMR': [83.00, 84.20, 76.00, 61.80, 70.00],
    'U5MR': [119.30, 126.60, 103.10, 82.60, 94.90],
    '%Under Nutrition': [53.50, 55.90, 47.30, 41.10, 47.00]
}

# Creating a pandas DataFrame
df = pd.DataFrame(data)

# Displaying the DataFrame
print(df)

# Plotting the data
plt.figure(figsize=(10, 6))

# Plotting IMR
plt.subplot(1, 3, 1)
plt.bar(df['Category'], df['IMR'])
plt.title('IMR')

# Plotting U5MR
plt.subplot(1, 3, 2)
plt.bar(df['Category'], df['U5MR'])
plt.title('U5MR')

# Plotting %Under Nutrition
plt.subplot(1, 3, 3)
plt.bar(df['Category'], df['%Under Nutrition'])
plt.title('%Under Nutrition')

# Adjusting layout
plt.tight_layout()

# Display the plots
plt.show()
```
